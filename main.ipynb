```python
import cv2
import numpy as np
import math
import pyautogui as pg
import pydirectinput


cam = cv2.VideoCapture(0)

cv2.namedWindow("game")


while True:
    ret, frame = cam.read()
    if not ret:
        print("failed to grab frame")
        break
        
    
    windowWidth=frame.shape[1]
    windowHeight=frame.shape[0]
    hsvFrame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    
    red_lower = np.array([0, 120, 20], np.uint8)
    red_upper = np.array([15, 255, 255], np.uint8)
    red_mask = cv2.inRange(hsvFrame, red_lower, red_upper)
    
    green_lower = np.array([45, 140, 20], np.uint8)
    green_upper = np.array([75, 255, 255], np.uint8)
    green_mask = cv2.inRange(hsvFrame, green_lower, green_upper)
    

    
    kernal = np.ones((5, 5), "uint8")
    
    red_mask = cv2.dilate(red_mask, kernal)
    res_red = cv2.bitwise_and(frame, frame,mask = red_mask)
    
    
    green_mask = cv2.dilate(green_mask, kernal)
    res_green = cv2.bitwise_and(frame, frame,mask = green_mask)
    

    
    contours, hierarchy = cv2.findContours(red_mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    
    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            red_rec = cv2.rectangle(frame, (x, y),(x + w, y + h),(0, 0, 255), 1)
            #cv2.putText(frame, "Red", (x, y),cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0, 0, 255),)
            mid_x=int(x+(w/2))
            mid_y=int(y+(h/2))
            
            mid_window_x=int(windowWidth/2)
            mid_window_y=int(windowHeight/2)
            cv2.line(frame, (mid_window_x,mid_window_y), (mid_x,mid_y), (255,0,0), 3)
            
            if mid_window_x-mid_x==0:
                angle=90
                
            else:
                slope=((mid_window_y-mid_y)/(mid_window_x-mid_x))
                angle=math.degrees(math.atan(slope))
                
            if(angle>0 and angle>20):
                pydirectinput.keyUp('left')
                
                
                pydirectinput.press('right',presses=2)
                
            if(angle>0 and angle<20):
                pydirectinput.keyUp('left')
                pydirectinput.keyDown('right')
            
                
                
            if(angle<0 and angle<-20):
                pydirectinput.keyUp('right')
                
                pydirectinput.press('left',presses=2)
                
                
            if(angle<0 and angle>-20):
                pydirectinput.keyUp('right')
                pydirectinput.keyDown('left')
            cv2.putText(frame, str(round(angle, 2)), (x+w, y+h),cv2.FONT_HERSHEY_SIMPLEX, 0.8,(0, 0, 255))
            
    
    contours, hierarchy = cv2.findContours(green_mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    
    for pic, contour in enumerate(contours):
        area = cv2.contourArea(contour)
        if(area > 300):
            x, y, w, h = cv2.boundingRect(contour)
            green_rec = cv2.rectangle(frame, (x, y),(x + w, y + h),(0, 255, 0), 1)
            
            mid_x=int(x+(w/2))
            mid_y=int(y+(h/2))
            
            cv2.line(frame, (mid_window_x-240,mid_window_y), (mid_x,mid_y), (255,0,0), 3)
            cv2.putText(frame, str(mid_window_y-mid_y),(x+w, y+h),cv2.FONT_HERSHEY_SIMPLEX, 0.8,(0, 255,0))
            
            if mid_window_y-mid_y>0:
                pydirectinput.keyUp('down')
                pydirectinput.press('up',presses=10)
                
            elif mid_window_y-mid_y<0:
                pydirectinput.keyUp('up')
                pydirectinput.keyDown('down')
                
          
      
    mid=int(windowHeight/2)
    
    cv2.line(frame, (0,mid), (windowWidth,mid), (255,0,0), 3)
    
        
    cv2.imshow("game", frame)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
    
    

cam.release()

cv2.destroyAllWindows()
```
